# Random walk

```python
def better(kassa, stavka, igra):
    n, p = 1, 0.5
    
    
    count_igr = 1 
    
    X = [0]
    Y = [kassa]
    
    while kassa >= stavka and count_igr<=igra:
        
        if stats.binom.rvs(n, p) == 1:
            kassa += stavka
            X.append(count_igr)
            Y.append(kassa)
        else:
            kassa -= stavka
            X.append(count_igr)
            Y.append(kassa)
        count_igr += 1
        
    plt.plot(X, Y)    
        
```

```python
plt.figure(figsize=(20,6))
plt.axhline(10000)
for i in range(10):
    better(10000,100,10000)
```

```python
import matplotlib.pyplot as plt
import numpy as np

def xy(r,phi):
    return r*np.cos(phi), r*np.sin(phi)


fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(111,aspect='equal')

x = np.arange(-30,30,0.1) 

phis=np.arange(-10,10,1)
r =10.

ax.scatter( *xy(r,phis), c='r' )
ax.plot(x, x**3)
ax.plot(x, -x**3)
ax.plot(x**3, x)
ax.plot(-x**3, x)

ax.plot(x, np.sin(x)*2)
ax.plot(-np.sin(x)*2, x)
ax.plot(x, np.cos(x)*2)
ax.plot(x+np.pi, np.cos(x)*2)
ax.plot(-np.cos(x)*2, x)
ax.plot(-np.cos(x)*2, x+np.pi)
ax.plot(x, np.tanh(x)**2)
ax.plot(x, -np.tanh(x)**2)
ax.plot(np.tanh(x)**2, x)
ax.plot(-np.tanh(x)**2, x)
ax.plot( *(0,0), *xy(r,phis))
ax.plot(x, x, c='blue')
ax.plot(x, x*-1, c='blue')
ax.axhline(0, c='black')
ax.axvline(0, c='black')
plt.xlim(-10,10)
plt.ylim(-10,10)
# plt.grid()
plt.show()
```
